## Encoding Categorical Features
### 1) One Hot Encoder

```python
from sklearn.preprocessing import OneHotEncoder

def hot_encoding(data):
    hot_encoder = OneHotEncoder()
    hot_protocols_matrix = hot_encoder.fit_transform(data['protocol_type'].values.reshape(-1,1)).toarray()
    hot_protocols_names = hot_encoder.get_feature_names_out(['protocol_type'])
    hot_encoded_protocols = pd.DataFrame(hot_protocols_matrix , columns=hot_protocols_names)
    data = pd.concat([data , hot_encoded_protocols] ,axis=1)
    data.drop('protocol_type' , axis=1 , inplace=True)
    return data
```

## Encoding Train & Test 'protocol_type' feature

```python
train_df = hot_encoding(train_df)
test_df = hot_encoding(test_df)
```

### 2) Binary Encoder

```python
from category_encoders import BinaryEncoder 

def bin_encoding(data):
    bin_encoder = BinaryEncoder()
    bin_encoded_features = bin_encoder.fit_transform(data['flag'])
    data = pd.concat([data , bin_encoded_features] , axis=1)
    data.drop('flag' , axis=1 , inplace=True)
    return data
```

## Importing Scikit-learn Lipraries 

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import f1_score
```

## Train Data preparation 

```python
x = train_df.drop('Class' , axis=1)
y = train_df['Class']

X_train, X_test, y_train, y_test = train_test_split(x , y , test_size=0.3 , random_state=42)
print (X_train.shape , X_test.shape ,y_train.shape , y_test.shape)
```

```python
test_df.shape
```

## Scaling Train Dataset

```python
scaler_train = MinMaxScaler()
X_tr_scaled = scaler_train.fit_transform(X_train)
X_te_scaled = scaler_train.transform(X_test)
X_tr_scaled.shape , X_te_scaled.shape
```

## Scaling Test Dataset

```python
test_df_scaled = scaler_train.transform(test_df)
test_df_scaled.shape
```

## Dimension Reduction using PCA
### PCA : Principal Component Analysis

```python
pca = PCA(n_components=20)
X_trainS = pca.fit_transform(X_tr_scaled)
X_testS = pca.transform(X_te_scaled)
expla_var = pca.explained_variance_ratio_
expla_var
```

```python
test_df_scaled = pca.transform(test_df_scaled)
```

## Models Training and Testing 

```python
def models (model_name , X_tr , y_tr , X_te , y_te):
    model_name.fit(X_tr , y_tr)
    y_predict = model_name.predict(X_te)
    acc = f1_score(y_te , y_predict , average='macro')
    return acc 
```

## KNN best key value & index

```python
f1_scores_list = []
for i in range(10):
    knn = models(KNeighborsClassifier(n_neighbors=i+1) , X_trainS  , y_train , X_testS , y_test)
    f1_scores_list.append(knn)
print ('KNN Maximum accuracy =' , max(f1_scores_list).round(5) , '& Key index =' , np.argmax(f1_scores_list)+1)
```

## Diffferent Models evaluating
### MACRO F1 Score 

```python
knn = models(KNeighborsClassifier(n_neighbors=3) , X_trainS  , y_train , X_testS , y_test)
svc = models(SVC()  ,  X_trainS  , y_train , X_testS , y_test )

rfc = models(RandomForestClassifier(max_depth=7 , random_state=42), X_trainS  , y_train , X_testS , y_test)
desTree = models(DecisionTreeClassifier(max_depth=7, random_state=42) , X_trainS  , y_train , X_testS , y_test)

logREG = models(LogisticRegression() , X_trainS  , y_train , X_testS , y_test)
gussianNB =  models(GaussianNB() , X_trainS  , y_train , X_testS , y_test )

print ('Key Neighbors Classifier =' , knn.round(5))
print ('Support Vector Machine   =' , svc.round(5))

print ('Random Forest Classifier =' , rfc.round(5))
print ('Decision Tree Classifier =' , desTree.round(5))

print ('Logistic Regression      =' , logREG.round(5))
print ('Gussian Naive Bayes      =' , gussianNB.round(5))
```
